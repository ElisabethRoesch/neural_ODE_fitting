```julia
# There are 5 sets of ERK/MEK measurements 
# 3 NGF: Cells are plated in medium containing a fixed amount of neuronal growth factor(5, 50, 100 NGF)
# 2 EGF: Epidermal growth factor (5, 50 EGF5)
```

```julia
using MAT, Statistics, Plots
```

```julia
vars = matread("/Users/eroesch/Dropbox/PhD/year_1/140728_data.mat");
keys(vars)
```

```julia
a = mean.(vars[ "NGF5_ppERK"])[1,:]
b = mean.(vars[ "NGF5_ppMEK"])[1,:]
scatter(1:26, a, label = "NGF5_ppERK")
scatter!(1:26, b, label = "NGF5_ppMEK")
```

```julia
plt=plot(a, b, xlab = "ppERK",ylab = "ppMEK")
scatter!(a,b)
```
